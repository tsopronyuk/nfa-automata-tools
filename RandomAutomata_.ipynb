```python
import random
from graphviz import Digraph
from IPython.display import display, Markdown, HTML, Image

def generate_random_alphabet():
    """Generates a random alphabet (2-3 symbols)"""
    letters = ['a', 'b', 'c', 'd', 'f', 'x', 'y', 'z']
    return random.sample(letters, random.randint(2, 3))

def generate_random_nfa(prefix, sigma=None, num_states=None):
    """Generates a random NFA with the given alphabet."""
    n = num_states if num_states else random.choice([3, 4])
    sigma = sigma if sigma else generate_random_alphabet()

    m = random.choice([2, 3]) if n > 2 else 2
    m = min(m, n)

    S = [f'{prefix}{i}' for i in range(1, n+1)]
    q0 = f'{prefix}1'
    F = random.sample(S, m)

    delta = {state: {symbol: set() for symbol in sigma} for state in S}

    # Ensuring reachability
    for i in range(2, n+1):
        qj = f'{prefix}{random.choice(range(1, i))}'
        s = random.choice(sigma)
        delta[qj][s].add(f'{prefix}{i}')

    # Adding random transitions
    for q in S:
        for s in sigma:
            for t in S:
                if t == q0:
                    continue
                if random.random() < 0.2:
                    delta[q][s].add(t)

    return S, sigma, delta, q0, F

def draw_nfa_for_print(S, sigma, delta, q0, F, title):
    """Creates a compact NFA graph for printing."""
    dot = Digraph(title, format='png')
    dot.attr(rankdir='LR',
             size='6,3',  # Reduced size
             margin='0.1',
             nodesep='0.3',
             ranksep='0.3',
             fontsize='9')

    dot.node('start', shape='none', width='0.15', height='0.15')
    dot.edge('start', q0, arrowsize='0.4')

    for state in S:
        if state in F:
            dot.node(state, shape='doublecircle', width='0.5', height='0.5', fixedsize='true')
        else:
            dot.node(state, shape='circle', width='0.5', height='0.5', fixedsize='true')

    for from_state in delta:
        for symbol in delta[from_state]:
            for to_state in delta[from_state][symbol]:
                dot.edge(from_state, to_state, label=symbol, arrowsize='0.4', fontsize='9')

    return dot

def display_two_nfas(nfa1, nfa2, op_name, titles=("A", "B")):
    """Displays two compact automata side by side."""
    # Using the alphabet of the first automaton for the second one
    nfa2 = (nfa2[0], nfa1[1], nfa2[2], nfa2[3], nfa2[4])

    dot1 = draw_nfa_for_print(*nfa1, f"{titles[0]} ({op_name})")
    dot2 = draw_nfa_for_print(*nfa2, f"{titles[1]} ({op_name})")

    display(HTML(f"""

        {dot1.pipe(format='svg').decode('utf-8')}
        {dot2.pipe(format='svg').decode('utf-8')}

    """))

def generate_variant(variant_num):
    """Generates an assignment variant with a forced page break."""
    # Start of a new variant - new page
    display(Markdown(f"Variant {variant_num}"))

    # Concatenation
    display(Markdown("### 1. Concatenation (A · B)"))
    sigma_conc = generate_random_alphabet()
    nfa1 = generate_random_nfa('q', sigma=sigma_conc)
    nfa2 = generate_random_nfa('p', sigma=sigma_conc)
    display_two_nfas(nfa1, nfa2, "Concatenation", ("A", "B"))

    display(Markdown("---"))

    # Union
    display(Markdown("### 2. Union (C ∪ D)"))
    sigma_union = generate_random_alphabet()
    nfa3 = generate_random_nfa('q', sigma=sigma_union)
    nfa4 = generate_random_nfa('p', sigma=sigma_union)
    display_two_nfas(nfa3, nfa4, "Union", ("C", "D"))

    display(Markdown("---"))

    # Iterations
    display(HTML("""

        3. Iteration (M*)
        4. Iteration (M+)

    """))

    sigma_iter = generate_random_alphabet()
    nfa5 = generate_random_nfa('q', sigma=sigma_iter)
    nfa6 = generate_random_nfa('p', sigma=sigma_iter)

    display_two_nfas(nfa5, nfa6, "Iteration", ("A*", "A+"))
    display(Markdown("---"))

# Generate 20 variants
for i in range(1, 21):
    generate_variant(i)
    # Additional break after the variant (if it didn't fit on one page)
    display(Markdown(""))
```

*2.1. Формальне визначення скінченних автоматів.*

Скінченний автомат (СА) — це математична модель дискретної системи, яка має скінченну кількість станів і переходить між ними під впливом вхідних символів. Формально, НКА визначається як п’ятірка:

A = (Q, Σ, δ, q₀, F),

де:
- Q — скінченна множина станів;
- Σ — скінченний алфавіт (множина вхідних символів);
- δ: Q × Σ → 2^Q — функція переходів, що відображає пару (стан, символ) у множину можливих наступних станів;
- q₀ ∈ Q — початковий стан;
- F ⊆ Q — множина фінальних (приймальних) станів.

На відміну від детермінованого скінченного автомата (ДКА), у НКА функція переходів може повертати декілька станів (включаючи порожню множину), що відображає неоднозначність поведінки. ДКА є окремим випадком НКА, де для кожної пари (стан, символ) функція переходів повертає не більше одного стану, тобто δ: Q × Σ → Q.

*2.2. Генератор випадкових НКА.*

Розроблений метод генерації випадкових НКА базується на принципі забезпечення досяжності всіх станів із початкового. Це важлива властивість для навчальних завдань, оскільки недосяжні стани не впливають на мову автомата, але можуть заплутувати студентів.

**Алгоритм 1. Генерація випадкового НКА**

*Вхід:* кількість станів n ∈ {3, 4, 5}, розмір алфавіту |Σ| ∈ {2, 3}, щільність переходів p = 0.2
*Вихід:* НКА A = (Q, Σ, δ, q₀, F)

1. Q ← {q₁, q₂, …, qₙ}
2. q₀ ← q₁
3. F ← випадковий вибір m станів з Q, де m ∈ [2, ⌊n/2⌋]
4. Σ ← випадковий вибір 2–3 символів з набору {a, b, c, d, f, x, y, z}
5. Ініціалізувати δ(q, a) ← ∅ для всіх q ∈ Q, a ∈ Σ
6. for i = 2 to n:
7.     qⱼ ← випадковий стан з {q₁, …, qᵢ₋₁}
8.     a ← випадковий символ з Σ
9.     δ(qⱼ, a) ← δ(qⱼ, a) ∪ {qᵢ}
10. for кожного q ∈ Q та a ∈ Σ:
11.     for кожного t ∈ Q, t ≠ q₀:
12.         if random() < p:
13.             δ(q, a) ← δ(q, a) ∪ {t}
14. return (Q, Σ, δ, q₀, F)

Ключовою особливістю алгоритму є гарантія досяжності всіх станів (кроки 6–9): для кожного стану qᵢ (i ≥ 2) додається перехід з деякого попереднього стану qⱼ (j < i), що гарантує існування шляху від q₀ до кожного стану. Це усуває необхідність додаткового етапу очищення автомата від недосяжних станів, який часто потрібен у загальних методах генерації.

Вибір параметрів n ∈ {3, 4, 5} та |Σ| ∈ {2, 3} зумовлений навчальними цілями: автомати з 3–5 станами є достатньо складними для демонстрації основних концепцій, але не надто громіздкими для ручної побудови студентами. Щільність переходів p = 0.2 забезпечує помірну кількість недетермінізму — достатньо цікавого для аналізу, але не надто складного.

*2.5. Алгоритм детермінізації.*

Важливою складовою курсу теорії автоматів є детермінізація НКА — перетворення недетермінованого автомата в еквівалентний детермінований. Цей процес демонструє фундаментальну властивість: класи мов, що розпізнаються НКА та ДКА, збігаються [1; 2].

У розробленій системі реалізовано алгоритм детермінізації, що базується на методі побудови підмножин (subset construction). Програмна реалізація у файлі NFA-to-DFA.ipynb виконує такі кроки:

1. **Зчитування вхідного НКА** з текстового файлу у форматі таблиці переходів. Перший рядок файлу містить алфавіт (символи, розділені символом «|»). Наступні рядки описують переходи для кожного стану. Символ «-» перед назвою стану позначає початковий стан, символ «*» — фінальний стан. Якщо стан є одночасно початковим і фінальним, використовується комбіноване позначення.

2. **Побудова множини станів ДКА** як підмножин станів вихідного НКА. Початковим станом ДКА є множина, що містить лише початковий стан НКА.

3. **Обчислення переходів** для кожної новоутвореної множини станів за кожним символом алфавіту. Для заданої множини станів S та символу a обчислюється об’єднання множин станів, досяжних з кожного стану S за символом a. Якщо отримана множина станів T ще не входила до множини станів ДКА, вона додається як новий стан.

4. **Визначення фінальних станів ДКА** — тими вважаються множини, що містять хоча б один фінальний стан вихідного НКА.

5. **Візуалізація отриманого ДКА** з використанням того ж модуля візуалізації, що й для вихідного НКА. Для зручності сприйняття стани ДКА позначаються літерами (A, B, C, …) замість множин станів НКА, що робить візуалізацію більш компактною та читабельною.

Важливою особливістю реалізації є **визначення досяжних станів** після побудови ДКА. Алгоритм аналізує, які стани ДКА є досяжними з початкового, та виключає недосяжні стани з фінального представлення. Це дозволяє отримати мінімальний за кількістю станів ДКА, еквівалентний вихідному НКА, що є важливим для наочності демонстрації.

*2.6.1. Тестування ШІ на різних форматах представлення.*

Для оцінки можливостей сучасних ШІ-моделей у розв’язуванні задач з теорії автоматів було проведено порівняльний аналіз з використанням ChatGPT-4 (версія від березня 2026 року). Типове завдання «Побудуйте НКА для конкатенації мов, заданих автоматами A та B» було подано двома способами:

- **Табличний формат** — автомати представлені у вигляді таблиць переходів, де рядки відповідають станам, стовпці — символам алфавіту, а на перетині зазначаються множини наступних станів.
- **Візуальний формат** — автомати представлені у вигляді діаграм переходів (рис. 1), де стани зображені колами, переходи — стрілками з мітками-символами.

Тестування проводилося на вибірці з 50 різних пар автоматів для кожного типу завдань (конкатенація, альтернатива, ітерація). Кожне завдання подавалося ШІ у відповідному форматі, а відповіді оцінювалися за критерієм коректності побудови результуючого НКА.

Результати експерименту показали, що у табличному форматі ШІ демонстрував високу ефективність: правильні відповіді було отримано для переважної більшості завдань для всіх типів операцій. Це підтверджує, що сучасні LLM добре справляються з формалізованими текстовими описами автоматів.

У візуальному форматі успішність ШІ була суттєво нижчою. Навіть при використанні інструментів для розпізнавання зображень, інтегрованих у мультимодальні версії моделей, правильне розпізнавання структури автомата за діаграмою переходів виявилося складним завданням. Успішність була значно нижчою навіть для найпростіших завдань. Це пояснюється тим, що візуальне представлення вимагає не лише розпізнавання графічних елементів (кола, стрілки, мітки), але й інтерпретації їх семантики як станів, переходів та символів.

Отримані результати підтверджують гіпотезу про те, що візуальне представлення автоматів є ефективним способом захисту від автоматичного розв’язування завдань штучним інтелектом. Студент, який отримує завдання у вигляді діаграм переходів, не може просто скопіювати його в ChatGPT — йому необхідно спочатку самостійно розпізнати структуру автомата, що вимагає глибокого розуміння матеріалу.

Опитування також показало, що студенти експериментальної групи оцінили завдання як більш цікаві та змістовні порівняно з контрольною групою. Це свідчить про те, що візуальний формат не лише ускладнює використання ШІ, але й підвищує залученість студентів у навчальний процес.

Таким чином, експериментальне дослідження підтверджує, що використання візуального формату представлення автоматів є ефективним засобом зниження частоти використання ШІ студентами при збереженні академічної успішності на прийнятному рівні.

*2.7. Порівняння з існуючими рішеннями.*

Розроблена система була порівняна з існуючими інструментами за такими критеріями: наявність генерації випадкових автоматів, якість візуалізації, підтримка парного відображення, можливість автоматизованої генерації варіантів завдань, реалізація детермінізації, захист від автоматичного розв’язування ШІ та інтеграція з Jupyter Notebook.

JFLAP [10] є потужним освітнім середовищем, що підтримує візуалізацію автоматів та інтерактивне виконання операцій. Однак він не має функціоналу автоматизованої генерації випадкових автоматів — кожен автомат потрібно вводити вручну, що робить створення великої кількості індивідуальних варіантів завдань надзвичайно трудомістким. Крім того, JFLAP є окремим додатком і не інтегрується в середовища типу Jupyter Notebook.

PySimpleAutomata [11] надає бібліотеку для роботи з автоматами в Python, включаючи реалізацію детермінізації. Однак її візуалізація обмежується форматом DOT, який вимагає додаткових інструментів для перетворення у зображення. Бібліотека не містить функціоналу для генерації випадкових автоматів, що обмежує її використання для створення великих наборів тестових прикладів. Крім того, представлення автоматів у текстовому форматі робить їх легко доступними для автоматичного розв’язування ШІ.

Розроблена система вирізняється унікальним поєднанням функціоналу:
- генерація випадкових НКА з гарантованою досяжністю станів та контрольованою щільністю переходів;
- високоякісна візуалізація у форматі SVG з налаштованими параметрами для друку;
- парне відображення автоматів для операцій конкатенації та альтернативи;
- автоматизована генерація 20 і більше індивідуальних варіантів завдань;
- реалізація детермінізації з визначенням досяжних станів та візуалізацією результатів;
- високий рівень захисту від автоматичного розв’язування ШІ завдяки використанню візуального формату;
- повна інтеграція з Jupyter Notebook, що дозволяє використовувати систему безпосередньо в навчальному середовищі.

**Висновки з дослідження і перспективи подальших розвідок у цьому напрямі.** У ході виконаної роботи було отримано такі наукові та практичні результати:

1. Розроблено метод генерації випадкових НКА із гарантованою досяжністю всіх станів та контрольованою щільністю переходів. Особливістю методу є забезпечення досяжності через додавання переходів від попередніх станів, що усуває необхідність додаткового етапу очищення автомата. Цей метод дозволяє створювати різноманітні навчальні завдання та експериментальні вибірки.

2. Реалізовано модуль візуалізації автоматів у вигляді діаграм переходів із налаштованими параметрами (розмір 6×3 дюйми, відстані між вузлами 0.3, шрифт 9 pt для ребер), що забезпечує компактне відображення, придатне для вставки в навчальні матеріали та друк на форматі A4.

3. Створено функцію парного відображення display_two_nfas(), яка забезпечує розміщення двох автоматів поруч з однаковими відступами та запобіганням розриву між сторінками. Це є критично важливим для візуалізації операцій конкатенації та альтернативи.

4. Розроблено генератор варіантів завдань generate_variant(), який автоматично формує індивідуальні варіанти для студентів. Кожен варіант містить три типи завдань: конкатенація, альтернатива та ітерація (A* та A+). Програма генерує 20 варіантів із автоматичним розміщенням на окремих сторінках.

5. Реалізовано алгоритм детермінізації НКА, що базується на методі побудови підмножин, з додатковим етапом визначення досяжних станів. Програмна реалізація (файл NFA-to-DFA.ipynb) дозволяє зчитувати НКА з текстового файлу у форматі таблиці переходів, будувати еквівалентний ДКА та візуалізувати результат з автоматичним перейменуванням станів для зручності сприйняття.

6. Проведено експериментальне дослідження, яке підтвердило:
   - сучасні ШІ-моделі (ChatGPT-4) демонструють високу успішність при розв’язуванні завдань у табличному форматі;
   - у візуальному форматі успішність ШІ є суттєво нижчою, що робить його ефективним засобом протидії;
   - використання візуального формату знижує частоту використання ШІ студентами при незначному зниженні академічної успішності;
   - студенти оцінюють завдання у візуальному форматі як більш цікаві та змістовні.

Наукова новизна отриманих результатів полягає в:
- розробці методу генерації випадкових НКА з гарантованою досяжністю станів, адаптованого для навчальних цілей;
- створенні комплексного інструментарію для автоматизованої генерації індивідуальних завдань, парної візуалізації та детермінізації НКА;
- експериментальному підтвердженні ефективності візуального представлення як засобу протидії використанню ШІ в навчальному процесі.

Практичне значення роботи полягає у створенні програмного забезпечення, яке дозволяє викладачам ефективно готувати індивідуальні завдання для студентів в умовах широкого доступу до інструментів ШІ, забезпечуючи об’єктивність оцінювання знань. Розроблені модулі можуть бути використані в навчальному процесі при вивченні курсів «Теорія формальних мов та автоматів» та «Мовні процесори» [3; 4; 5]. Програмна реалізація описаних алгоритмів доступна у відкритому репозиторії за посиланням [URL].

Перспективи подальших розвідок у цьому напрямі включають:
1. Розширення функціоналу для підтримки інших типів завдань (мінімізація ДКА, перетворення регулярних виразів, перетин та доповнення автоматів).
2. Інтеграцію з системами дистанційного навчання (Moodle, Google Classroom) для автоматичного розподілу варіантів, збору відповідей та автоматизованої перевірки.
3. Розробку методів автоматичної перевірки виконаних завдань, представлених у візуальному форматі, з використанням технік комп’ютерного зору.
4. Дослідження можливостей адаптації ШІ до розпізнавання діаграм переходів та розробку відповідних контрзаходів для збереження ефективності запропонованого підходу.
5. Додавання можливості експорту в різні формати (PDF, LaTeX, HTML) для інтеграції з різними системами документообігу та видавничими платформами.
